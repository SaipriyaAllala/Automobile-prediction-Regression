```python
import pandas as pd
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
data=pd.read_csv(r"C:\Users\saipr\Downloads\Automobile.csv")
```

```python
num=data.select_dtypes(['int64','float64'])
```

```python
num.columns
```

```python
num_cols=num.columns.tolist()
```

```python
from sklearn.preprocessing import StandardScaler

std_scale = StandardScaler()

std_scale
```

```python
for i in num_cols:
    num[i]= std_scale.fit_transform(num[i].values.reshape(-1,1))
```

```python
num
```

```python
obj=data.select_dtypes('object')
```

```python
obj
```

```python
obj_cols=obj.columns.tolist()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
for i in obj_cols:
    obj[i]=labelencoder.fit_transform(obj[i].astype('str'))
```

```python
obj
```

```python
new_data=pd.concat([num,obj],axis=1)
```

```python
new_data
```

```python
x=new_data.drop('price',axis=1)
y=new_data['price']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=40)
 
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
rm= LinearRegression()
rm.fit(x_train,y_train)
```

```python
for idx ,col_name in enumerate(x_train.columns):
    print("The Coefficient for {} is {}".format(col_name,rm.coef_[idx]))
```

```python
rm.score(x_test,y_test)
```
